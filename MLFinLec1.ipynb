Reading the Excel Files:

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns

# Make plots larger
plt.rcParams['figure.figsize'] = (15, 9)

#Loading all the csv files

df_usch = pd.read_excel("dexchus.xls")
df_useu = pd.read_excel("dexuseu.xls")
df_usjp = pd.read_excel("dexjpus.xls")
```

Steps fro Data Cleaning:
    1. Drop the rows with Unnecessary data
    2. Drop rows with value = 0 for certain days

```python
#Removing columns with no data
dfusjp=df_usjp[10:]
```

```python
dfusjp.head()
```

```python
#Giving header names
dfusjp.columns = ['observation_date', 'value']
```

```python
dfusjp.head()
```

```python
#Drop rows with values of 0
dfusjp = dfusjp[dfusjp.value != 0]
```

```python
dfusjp.head()
```

Repeat the process above for US-EU and US_CH datasets

```python
dfuseu=df_useu[10:]
```

```python
dfuseu.columns = ['observation_date', 'value']
dfuseu = dfuseu[dfuseu.value != 0]
dfuseu.head()
```

```python
dfusch=df_usch[10:]
```

```python
dfusch.columns = ['observation_date', 'value']
dfusch = dfusch[dfusch.value != 0]
dfusch.head()
```
